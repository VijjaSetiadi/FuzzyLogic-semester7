# SOAL A: FUZZY LOGIC MAMDANI - DIAGNOSA TINGKAT RISIKO DBD


```python
import numpy as np

# ==================================================================================
# SOAL A: FUZZY LOGIC MAMDANI - DIAGNOSA TINGKAT RISIKO DBD
# ==================================================================================

print("="*80)
print(" SOAL A: METODE FUZZY LOGIC (MAMDANI) ".center(80, "="))
print(" Diagnosa Tingkat Risiko Demam Berdarah ".center(80))
print("="*80)

# Data Pasien dari soal (Contoh Kasus 4)
suhu_tubuh = 38.5  # °C
trombosit = 120    # x10³/L
leukosit = 3.8     # x10³/L

print(f"\n4. Contoh Kasus Pasien:")
print(f"   Parameter          Nilai")
print(f"   Suhu Tubuh         {suhu_tubuh}°C")
print(f"   Trombosit          {trombosit} x10³/L")
print(f"   Leukosit           {leukosit} x10³/L")

# A. INPUT - Fungsi Keanggotaan

# 1. Suhu Tubuh (°C)
def suhu_normal(x):
    """Normal: 36 - 37.5"""
    if x <= 36:
        return 0
    elif 36 < x <= 36.75:
        return (x - 36) / 0.75
    elif 36.75 < x <= 37.5:
        return (37.5 - x) / 0.75
    else:
        return 0

def suhu_demam(x):
    """Demam: 37 - 39"""
    if x <= 37:
        return 0
    elif 37 < x <= 38:
        return (x - 37) / 1
    elif 38 < x <= 39:
        return (39 - x) / 1
    else:
        return 0

def suhu_tinggi(x):
    """Tinggi: 38 - 41"""
    if x <= 38:
        return 0
    elif 38 < x <= 39.5:
        return (x - 38) / 1.5
    elif x > 39.5:
        return 1
    else:
        return 0

# 2. Jumlah Trombosit (10³/L)
def trombosit_rendah(x):
    """Rendah: < 150"""
    if x >= 150:
        return 0
    elif 100 < x < 150:
        return (150 - x) / 50
    else:
        return 1

def trombosit_normal(x):
    """Normal: 150 - 400"""
    if x <= 150:
        return 0
    elif 150 < x <= 275:
        return (x - 150) / 125
    elif 275 < x <= 400:
        return (400 - x) / 125
    else:
        return 0

# 3. Jumlah Leukosit (10³/L)
def leukosit_rendah(x):
    """Rendah: < 4"""
    if x >= 4:
        return 0
    elif 2 < x < 4:
        return (4 - x) / 2
    else:
        return 1

def leukosit_normal(x):
    """Normal: 4 - 10"""
    if x <= 4:
        return 0
    elif 4 < x <= 7:
        return (x - 4) / 3
    elif 7 < x <= 10:
        return (10 - x) / 3
    else:
        return 0

# Fuzzifikasi Input
mu_suhu_normal = suhu_normal(suhu_tubuh)
mu_suhu_demam = suhu_demam(suhu_tubuh)
mu_suhu_tinggi = suhu_tinggi(suhu_tubuh)

mu_trombosit_rendah = trombosit_rendah(trombosit)
mu_trombosit_normal = trombosit_normal(trombosit)

mu_leukosit_rendah = leukosit_rendah(leukosit)
mu_leukosit_normal = leukosit_normal(leukosit)

print(f"\nFuzzifikasi:")
print(f"   Suhu - Normal: {mu_suhu_normal:.4f}, Demam: {mu_suhu_demam:.4f}, Tinggi: {mu_suhu_tinggi:.4f}")
print(f"   Trombosit - Rendah: {mu_trombosit_rendah:.4f}, Normal: {mu_trombosit_normal:.4f}")
print(f"   Leukosit - Rendah: {mu_leukosit_rendah:.4f}, Normal: {mu_leukosit_normal:.4f}")

# B. OUTPUT - Fungsi Keanggotaan Tingkat Risiko DBD
def risiko_rendah(x):
    """Rendah: 0 - 40"""
    if x >= 40:
        return 0
    elif 20 < x < 40:
        return (40 - x) / 20
    else:
        return 1

def risiko_sedang(x):
    """Sedang: 30 - 70"""
    if x <= 30:
        return 0
    elif 30 < x <= 50:
        return (x - 30) / 20
    elif 50 < x <= 70:
        return (70 - x) / 20
    else:
        return 0

def risiko_tinggi(x):
    """Tinggi: 60 - 100"""
    if x <= 60:
        return 0
    elif 60 < x < 80:
        return (x - 60) / 20
    else:
        return 1

# 3. Rule Fuzzy (Inferensi)
print(f"\n3. Rule Fuzzy:")

# Rule 1: Jika Suhu tinggi AND Trombosit rendah AND Leukosit rendah maka Risiko = Tinggi
rule1 = min(mu_suhu_tinggi, mu_trombosit_rendah, mu_leukosit_rendah)
print(f"   Rule 1: min({mu_suhu_tinggi:.4f}, {mu_trombosit_rendah:.4f}, {mu_leukosit_rendah:.4f}) = {rule1:.4f} → Risiko = Tinggi")

# Rule 2: Jika Suhu demam AND Trombosit rendah maka Risiko = Tinggi
rule2 = min(mu_suhu_demam, mu_trombosit_rendah)
print(f"   Rule 2: min({mu_suhu_demam:.4f}, {mu_trombosit_rendah:.4f}) = {rule2:.4f} → Risiko = Tinggi")

# Rule 3: Jika Suhu demam AND Trombosit normal maka Risiko = Sedang
rule3 = min(mu_suhu_demam, mu_trombosit_normal)
print(f"   Rule 3: min({mu_suhu_demam:.4f}, {mu_trombosit_normal:.4f}) = {rule3:.4f} → Risiko = Sedang")

# Rule 4: Jika Suhu normal AND Trombosit normal AND Leukosit normal maka Risiko = Rendah
rule4 = min(mu_suhu_normal, mu_trombosit_normal, mu_leukosit_normal)
print(f"   Rule 4: min({mu_suhu_normal:.4f}, {mu_trombosit_normal:.4f}, {mu_leukosit_normal:.4f}) = {rule4:.4f} → Risiko = Rendah")

# Rule 5: Jika Suhu tinggi AND Leukosit normal maka Risiko = Sedang
rule5 = min(mu_suhu_tinggi, mu_leukosit_normal)
print(f"   Rule 5: min({mu_suhu_tinggi:.4f}, {mu_leukosit_normal:.4f}) = {rule5:.4f} → Risiko = Sedang")

# Agregasi (MAX untuk output yang sama)
mu_tinggi = max(rule1, rule2)
mu_sedang = max(rule3, rule5)
mu_rendah = rule4

print(f"\nAgregasi Output:")
print(f"   Risiko Tinggi: max({rule1:.4f}, {rule2:.4f}) = {mu_tinggi:.4f}")
print(f"   Risiko Sedang: max({rule3:.4f}, {rule5:.4f}) = {mu_sedang:.4f}")
print(f"   Risiko Rendah: {mu_rendah:.4f}")

# Defuzzifikasi (Metode Centroid/Center of Gravity)
x = np.linspace(0, 100, 1000)

# Hitung membership untuk setiap output yang dipotong oleh rule strength
output_rendah = np.minimum(mu_rendah, [risiko_rendah(xi) for xi in x])
output_sedang = np.minimum(mu_sedang, [risiko_sedang(xi) for xi in x])
output_tinggi = np.minimum(mu_tinggi, [risiko_tinggi(xi) for xi in x])

# Agregasi final (MAX)
aggregated = np.maximum(output_rendah, np.maximum(output_sedang, output_tinggi))

# Centroid
if np.sum(aggregated) == 0:
    tingkat_risiko = 0
else:
    tingkat_risiko = np.sum(x * aggregated) / np.sum(aggregated)

# Tentukan kategori
if tingkat_risiko < 40:
    kategori = "RENDAH"
elif 40 <= tingkat_risiko < 70:
    kategori = "SEDANG"
else:
    kategori = "TINGGI"

print(f"\n{'='*80}")
print(f"HASIL DIAGNOSA RISIKO DBD:")
print(f"   Tingkat Risiko: {tingkat_risiko:.2f}/100")
print(f"   Kategori: {kategori}")
print(f"{'='*80}\n")
```

# SOAL B: ANFIS - PREDIKSI PRODUKSI JAGUNG

```python
# ==================================================================================
# SOAL B: ANFIS - PREDIKSI PRODUKSI JAGUNG
# ==================================================================================

print("\n" + "="*80)
print(" SOAL B: METODE ANFIS (ADAPTIVE NEURO-FUZZY INFERENCE SYSTEM) ".center(80, "="))
print(" Prediksi Produksi Jagung ".center(80))
print("="*80)

# Data input dari soal (Contoh kasus: x1 = 60, x2 = 26)
x1 = 60  # soil_moisture (%)
x2 = 26  # avg_temp (°C)

print(f"\nContoh kasus, Hitunglah prediksi target jagung untuk data x1 = {x1}, x2 = {x2}.")
print(f"\nData Input:")
print(f"   soil_moisture (x1): {x1}%")
print(f"   avg_temp (x2): {x2}°C")

# 1) Struktur Model (sederhana)

# Fungsi keanggotaan untuk Input 1: soil_moisture (x1) — domain = [0,100]
def x1_mf_low(x):
    """MF Low untuk x1: trimf(0,30,60)"""
    if x <= 0:
        return 1
    elif 0 < x <= 30:
        return 1
    elif 30 < x <= 60:
        return (60 - x) / 30
    else:
        return 0

def x1_mf_high(x):
    """MF High untuk x1: trimf(40,70,100)"""
    if x <= 40:
        return 0
    elif 40 < x <= 70:
        return (x - 40) / 30
    elif 70 < x <= 100:
        return 1
    else:
        return 1

# Fungsi keanggotaan untuk Input 2: avg_temp (x2) — domain = [15,35]
def x2_mf_low(x):
    """MF Low untuk x2: trimf(15,20,25)"""
    if x <= 15:
        return 1
    elif 15 < x <= 20:
        return 1
    elif 20 < x <= 25:
        return (25 - x) / 5
    else:
        return 0

def x2_mf_high(x):
    """MF High untuk x2: trimf(22,28,35)"""
    if x <= 22:
        return 0
    elif 22 < x <= 28:
        return (x - 22) / 6
    elif 28 < x <= 35:
        return 1
    else:
        return 1

# Fuzzifikasi
mu_x1_low = x1_mf_low(x1)
mu_x1_high = x1_mf_high(x1)
mu_x2_low = x2_mf_low(x2)
mu_x2_high = x2_mf_high(x2)

print(f"\n1) Struktur model (sederhana)")
print(f"   Derajat Keanggotaan:")
print(f"      x1 IS Low : {mu_x1_low:.4f}")
print(f"      x1 IS High: {mu_x1_high:.4f}")
print(f"      x2 IS Low : {mu_x2_low:.4f}")
print(f"      x2 IS High: {mu_x2_high:.4f}")

# Gunakan Sugeno first-order: setiap rule i memiliki consequent linear
# yi = pi*x1 + qi*x2 + ri

# Parameter (dikelompokkan per rule) - dari soal
params = {
    'R1': {'p': 0.0692, 'q1': 0.0461, 'r1': 0.00231},
    'R2': {'p': 0.0698, 'q2': 0.0390, 'r2': 0.00139},
    'R3': {'p': 0.0585, 'q3': 0.0183, 'r3': 0.00084},
    'R4': {'p': 0.0505, 'q4': 0.0318, 'r4': 0.00115}
}

print(f"\n   Parameter (dikelompokkan per rule):")
print(f"      Rule 1 (R1): p1 = {params['R1']['p']}, q1 = {params['R1']['q1']}, r1 = {params['R1']['r1']}")
print(f"      Rule 2 (R2): p2 = {params['R2']['p']}, q2 = {params['R2']['q2']}, r2 = {params['R2']['r2']}")
print(f"      Rule 3 (R3): p3 = {params['R3']['p']}, q3 = {params['R3']['q3']}, r3 = {params['R3']['r3']}")
print(f"      Rule 4 (R4): p4 = {params['R4']['p']}, q4 = {params['R4']['q4']}, r4 = {params['R4']['r4']}")

# Aturan (label):
print(f"\n   Aturan (label):")

# 1. R1: IF x1 IS Low AND x2 IS Low → y1 = p1*x1 + q1*x2 + r1
w1 = min(mu_x1_low, mu_x2_low)
y1 = params['R1']['p'] * x1 + params['R1']['q1'] * x2 + params['R1']['r1']
print(f"      1. R1: IF x1 IS Low AND x2 IS Low → y1 = p1*x1 + q1*x2 + r1")
print(f"         w1 = min({mu_x1_low:.4f}, {mu_x2_low:.4f}) = {w1:.4f}")
print(f"         y1 = {params['R1']['p']}*{x1} + {params['R1']['q1']}*{x2} + {params['R1']['r1']} = {y1:.5f}")

# 2. R2: IF x1 IS Low AND x2 IS High → y2 = p2*x1 + q2*x2 + r2
w2 = min(mu_x1_low, mu_x2_high)
y2 = params['R2']['p'] * x1 + params['R2']['q2'] * x2 + params['R2']['r2']
print(f"      2. R2: IF x1 IS Low AND x2 IS High → y2 = p2*x1 + q2*x2 + r2")
print(f"         w2 = min({mu_x1_low:.4f}, {mu_x2_high:.4f}) = {w2:.4f}")
print(f"         y2 = {params['R2']['p']}*{x1} + {params['R2']['q2']}*{x2} + {params['R2']['r2']} = {y2:.5f}")

# 3. R3: IF x1 IS High AND x2 IS Low → y3 = p3*x1 + q3*x2 + r3
w3 = min(mu_x1_high, mu_x2_low)
y3 = params['R3']['p'] * x1 + params['R3']['q3'] * x2 + params['R3']['r3']
print(f"      3. R3: IF x1 IS High AND x2 IS Low → y3 = p3*x1 + q3*x2 + r3")
print(f"         w3 = min({mu_x1_high:.4f}, {mu_x2_low:.4f}) = {w3:.4f}")
print(f"         y3 = {params['R3']['p']}*{x1} + {params['R3']['q3']}*{x2} + {params['R3']['r3']} = {y3:.5f}")

# 4. R4: IF x1 IS High AND x2 IS High → y4 = p4*x1 + q4*x2 + r4
w4 = min(mu_x1_high, mu_x2_high)
y4 = params['R4']['p'] * x1 + params['R4']['q4'] * x2 + params['R4']['r4']
print(f"      4. R4: IF x1 IS High AND x2 IS High → y4 = p4*x1 + q4*x2 + r4")
print(f"         w4 = min({mu_x1_high:.4f}, {mu_x2_high:.4f}) = {w4:.4f}")
print(f"         y4 = {params['R4']['p']}*{x1} + {params['R4']['q4']}*{x2} + {params['R4']['r4']} = {y4:.5f}")

# Weighted average (defuzzification Sugeno)
total_weight = w1 + w2 + w3 + w4

if total_weight == 0:
    output = 0
else:
    output = (w1 * y1 + w2 * y2 + w3 * y3 + w4 * y4) / total_weight

print(f"\n   Defuzzification (Weighted Average):")
print(f"      Total Weight = {w1:.4f} + {w2:.4f} + {w3:.4f} + {w4:.4f} = {total_weight:.4f}")
print(f"      Output = ({w1:.4f}*{y1:.5f} + {w2:.4f}*{y2:.5f} + {w3:.4f}*{y3:.5f} + {w4:.4f}*{y4:.5f}) / {total_weight:.4f}")
print(f"      Output = {output:.5f}")

print(f"\n{'='*80}")
print(f"HASIL PREDIKSI PRODUKSI JAGUNG:")
print(f"   Prediksi Target Jagung: {output:.5f} ton/ha")
print(f"{'='*80}\n")
print("\n" + "="*80)
print("="*80)
```
